# Generalised framework for PDEs 

Instead of individually deriving each PDE as we did in Section 4.1 , we look to Brown [1] for a general framework for PDEs to approximate Asian option prices. In this Section, we recover the significant results of Brown and derive some more PDEs to run a performance comparison.

assuming that $t_0 = 0,$ we know that the payoff of the fixed strike asian option is given as $max(\frac{1}{T}\int_{0}^{T} S_u \,du - K, 0)$ where we define the average asset price $A_t = q(t)S_t + \frac{e^{-r(T-t)}}{T}\int_{0}^{T} S_u \,du$. Note that this is not the same $A_t$ as the riskless asset. the function $q(t)$ is given by $q(t) = \left\{ \begin{array}{ll} \frac{1-e^{-r(T-t)}}{rT} \quad r \neq \delta \\ \frac{T-t}{T} \quad r = \delta \end{array} \right.$ where we assume the dividend yield $\delta = 0$. Lastly, Brown sets $$\epsilon_t = \frac{a(t)S_t + b(t) \left[ A_t - Ke^{-r(T-t)}\right]}{S_t}$$ 

where $a(t)$ and $b(t)$ are arbitrary functions and $b(t)$ is non zero. For this $\epsilon_t$ function, we have that the value of the Asian option at time t is given by $S_t \phi(t,\epsilon_t)$ where $\phi$ satisfies the partial differential equation $$\frac{\partial}{\partial t}\phi(t,\epsilon_t) + \left[ a'(t) + b'(t) \frac{(\epsilon - a(t))}{b(t)} \right] \frac{\partial}{\partial \epsilon}\phi(t,\epsilon_t) + \frac{1}{2}\sigma^{2}\left[ a(t) + b(t)q(t) - \epsilon \right]^{2} \frac{\partial^2}{\partial \epsilon^2}\phi(t,\epsilon_t) = 0$$

In fact, we can recover Roger & Shi's PDE, Vecer's New PDE and the Dubois-Lelievre PDE using this framework as shown by [1].:

**5.1 ROGER & SHI'S PDE**

Take $a(t) = q(t)e^{(r-\delta)(T-t)}, \quad b(t) = -e^{(r-\delta)(T-t)}$

Thus we have $\epsilon_t = q(t)e^{r(T-t)} - e^{r(T-t)}\frac{A_t - Ke^{-\delta(T-t)}}{S_t}$

Note that $\epsilon_t$ here is equivalent to $Z_t$ in (2)

$a'(t) = -(r-\delta)e^{(r-\delta)(T-t)}q(t) - \frac{1}{T}, \quad b'(t) =(r-\delta)e^{(r-\delta)(T-t)}$

Thus we have $\left[ a'(t) + b'(t) \frac{(\epsilon - a(t))}{b(t)} \right] = -\left((r-\delta)e^{(r-\delta)(T-t)}q(t) + \frac{1}{T} \right) - \frac{(r-\delta)e^{(r-\delta)(T-t)}(\epsilon_t -a(t))}{e^{(r-\delta)(T-t)}} =$  $-\frac{1}{T} - (r-\delta)\epsilon_t = -\left(\frac{1}{T} + r\epsilon_t \right)$

And $\left[ a(t) + b(t)q(t) - \epsilon \right]^{2} = \left( q(t)e^{(r-\delta)(T-t)} - q(t)e^{(r-\delta)(T-t)}- \epsilon \right)^2 = \epsilon^2$

Thus by the general framework, we recover Roger & Shi's PDE as:
$$\frac{\partial}{\partial t}\phi(t,\epsilon_t) - \left(r\epsilon_t + \frac{1}{T} \right)\frac{\partial}{\partial \epsilon}\phi(t,\epsilon_t) + \frac{1}{2} \sigma^{2}\epsilon^{2} \frac{\partial^2}{\partial \epsilon^2}\phi(t,\epsilon_t) = 0 \quad ... (4)$$

**NOTE**: take note here that the coefficient of $\frac{1}{T}\frac{\partial}{\partial \epsilon}\phi(t,\epsilon_t)$ is negative and not positive as in equation (2)

**5.2 VECER'S NEW PDE**

Take $a(t) = 0, \quad b(t) = 1$

Thus we have $\epsilon_t = \frac{A_t - Ke^{-r(T-t)}}{S_t}$

we evaluate: $\left[ a'(t) + b'(t) \frac{(\epsilon - a(t))}{b(t)} \right] = 0$ since $a'(t) = b'(t) = 0$

And $\left[ a(t) + b(t)q(t) - \epsilon \right]^{2} = (q(t) - \epsilon)^2$

Thus by the generalised framework we recover Vecer's new PDE as: 

$$\frac{\partial}{\partial t}\phi(t,\epsilon_t) + \frac{1}{2}\sigma^2 (q(t)-\epsilon)^2 \frac{\partial^2}{\partial \epsilon^2}\phi (t, \epsilon_t) = 0 \quad ... (5)$$

Note that this is not the same PDE as (3) due to the absence of the $\frac{\partial}{\partial \epsilon}\phi(t,\epsilon_t)$ term in this equation. Hence we refer to this as Vecer's NEW PDE.

**5.3 DUBOIS-LELIEVRE'S PDE**

Take $a(t) = \frac{t}{T} - q(t)b(t), \quad b(t) = e^{(r-\delta)(T-t)}$

Thus we have $\epsilon_t = \frac{t}{T} + q(t)e^{(r-\delta)(T-t)} - e^{(r-\delta)(T-t)}\frac{A_t - Ke^{-r(T-t)}}{S_t}$

$a'(t) = -q(t)(r-\delta)e^{(r-\delta)(T-t)}, \quad b'(t) = -(r-\delta)e^{(r-\delta)(T-t)}$

we evaluate:  $\left[ a'(t) + b'(t) \frac{(\epsilon - a(t))}{b(t)} \right] = -q(t)(r-\delta)e^{(r-\delta)(T-t)} -(r-\delta)e^{(r-\delta)(T-t)} \left[ \frac{\epsilon - \frac{t}{T} - q(t)e^{(r-\delta)(T-t)}}{e^{(r-\delta)(T-t)}} \right] = (r-\delta) \left[ -q(t)e^{(r-\delta)(T-t)} + e^{(r-\delta)(T-t)}\frac{A_t - Ke^{-r(T-t)}}{S_t} \right] = (r-\delta)\left[ \frac{t}{T} - \epsilon_t \right]$ 

And $\left[ a(t) + b(t)q(t) - \epsilon \right]^{2} = \left[ \frac{t}{T} - q(t)b(t) + q(t)b(t) - \epsilon \right]^2 = \left[ \frac{t}{T} - \epsilon \right]^2$

Thus by the generalised framework we recover Dubois-Lelievre's PDE as:

$$\frac{\partial}{\partial t}\phi(t,\epsilon_t) + (r-\delta)\left( \frac{t}{T} - \epsilon_t \right)\frac{\partial}{\partial \epsilon}\phi(t,\epsilon_t) + \frac{1}{2}\sigma^2 \left( \frac{t}{T} - \epsilon \right)^2 \frac{\partial^2}{\partial \epsilon^2}\phi (t, \epsilon_t) = 0 \quad ... (6)$$

Note that the Dubois-Lelievre PDE is similar to Vecer's PDE but in this case, stock position $q_t = \frac{t}{T}$


We can now derive Vecer's (original) PDE and Roger Lord's PDE by this method too:

**5.5 VECER'S PDE**

Take $a(t) = 1 - \frac{t}{T} - q(t)b(t), \quad b(t) = e^{(r-\delta)(T-t)}$

Thus we have $\epsilon_t = 1 - \frac{t}{T} - q(t)e^{(r-\delta)(T-t)} + \frac{e^{(r-\delta)(T-t)}\left[ A_t - Ke^{-r(T-t)} \right]}{S_t}$

we have $a'(t) = \frac{e^{(r-\delta)(T-t)}-1}{T}, \quad b'(t) = -(r-\delta)e^{(r-\delta)(T-t)}$

Thus we get:

$\left[ a'(t) + b'(t) \frac{(\epsilon - a(t))}{b(t)} \right] = \frac{e^{(r-\delta)(T-t)}-1}{T} - (r-\delta)e^{(r-\delta)(T-t)} \left( \frac{\epsilon -1 + \frac{t}{T} + q(t)e^{(r-\delta)(T-t)}}{e^{(r-\delta)(T-t)}} \right) = \frac{e^{(r-\delta)(T-t)}-1}{T} - (r-\delta)\left( \epsilon -1 + \frac{t}{T} + \frac{e^{(r-\delta)(T-t)}-1}{(r-\delta)T} \right) = r \left( 1 - \frac{t}{T} - \epsilon \right)$

And $\left[ a(t) + b(t)q(t) - \epsilon \right]^{2} = \left[ 1 - \frac{t}{T} - q(t)b(t) + q(t)b(t) -\epsilon \right]^{2} = \left[ 1 - \frac{t}{T} - \epsilon \right]^2$

Thus by the generalised framework we recover Vecer's PDE as:

$$\frac{\partial}{\partial t}\phi(t,\epsilon_t) + r\left( 1 -\frac{t}{T} - \epsilon_t \right)\frac{\partial}{\partial \epsilon}\phi(t,\epsilon_t) + \frac{1}{2}\sigma^2 \left(1 - \frac{t}{T} - \epsilon_t \right)^2 \frac{\partial^2}{\partial \epsilon^2}\phi (t, \epsilon) = 0 \quad ...(8)$$

**5.6 ROGER LORD'S PDE**

Take $a(t) = 1 - q(t)b(t), \quad b(t) = e^{(r-\delta)(T-t)}$

We have $\epsilon_t = 1- \frac{e^{(r-\delta)(T-t)} -1}{rT} + e^{(r-\delta)(T-t)}\frac{\left[A_t - Ke^{-r(T-t)} \right]}{S_t}$

And we have: $a'(t) = \frac{e^{(r-\delta)(T-t)}-1}{T} + \frac{1}{T}, \quad b'(t) = -(r-\delta)e^{(r-\delta)(T-t)}$

Thus we evaluate:

$\left[ a'(t) + b'(t) \frac{(\epsilon - a(t))}{b(t)} \right] = \frac{e^{(r-\delta)(T-t)}-1}{T} - (r-\delta)e^{(r-\delta)(T-t)}\left( \frac{\epsilon - 1 + \frac{e^{(r-\delta)(T-t)}-1}{rT}}{e^{(r-\delta)(T-t)}} \right) = \frac{e^{(r-\delta)(T-t)}-1}{T} - r \left( \epsilon - 1 + \frac{e^{(r-\delta)(T-t)}-1}{rT} \right) = r(1-\epsilon)$

$\left[ a(t) + b(t)q(t) - \epsilon \right]^{2} = \left[ 1 - b(t)q(t) + b(t)q(t) - \epsilon \right]^2 = \left(1 - \epsilon \right)^2$

thus by the generalised framework we recover Roger Lord's PDE as:

$$\frac{\partial}{\partial t}\phi(t,\epsilon_t) + r(1-\epsilon)\frac{\partial}{\partial \epsilon}\phi(t,\epsilon_t) + \frac{1}{2}\sigma^2 (1-\epsilon)^2\frac{\partial^2}{\partial \epsilon^2}\phi (t, \epsilon) = 0 \quad ... (9)$$






# Numerical Methods to solve Partial Differential Equations

**Finite difference methods to solve a PDE**

Using finite difference mesh: for a function $f(t,S)$ divide the $(t,S)$ plane into discrete points

S axis is divided into $\vartriangle S$ equal parts and t axis is divided into $\vartriangle t$ equal parts.

the Mesh points can be written as $(i \vartriangle t, j \vartriangle S)$ where $i = 0,1, ... ,N$ and $j = 0,1, ... ,M$

**Forward, backward and central difference approximation to 1st order and 2nd order derivatives:**

Forward: $\frac{\partial f (t,S)}{\partial t} \approx \frac{f(t+\vartriangle t, S) - f(t,S)}{\vartriangle t}$

Backward: $\frac{\partial f (t,S)}{\partial t} \approx \frac{f(t,S) - f(t - \vartriangle t, S)}{\vartriangle t}$

Central: $\frac{\partial f (t,S)}{\partial t} \approx \frac{f(t+\vartriangle t, S) - f(t-\vartriangle t,S)}{2 \vartriangle t}$

we use the following notation:
$f(t,S)$ at points $(i \vartriangle t, j \vartriangle S) = f_{i,j} = f(i \vartriangle t, j \vartriangle S)$

thus we have

Forward difference: $\frac{\partial f}{\partial t} \approx \frac{f_{i+1,j} - f_{i,j}}{\vartriangle t} \quad \frac{\partial f}{\partial s} \approx \frac{f_{i,j+1}-f_{i,j}}{\vartriangle S}$

Backward difference: $\frac{\partial f}{\partial t} \approx \frac{f_{i,j}-f_{i-1,j}}{\vartriangle t} \quad \frac{\partial f}{\partial s} \approx \frac{f_{i,j} - f_{i,j-1}}{\vartriangle S}$

Central difference: $\frac{\partial f}{\partial t} \approx \frac{f_{i+1,j} - f_{i-1,j}}{2 \vartriangle t} \quad \frac{\partial f}{\partial S} \approx \frac{f_{i,j+1} - f_{i,j-1}}{2 \vartriangle S}$

Second derivative: $\frac{\partial^{2}f}{\partial S^{2}} \approx \frac{\frac{f_{i,j+1} - f_{i,j}}{\vartriangle S} - \frac{f_{i,j} - f{i,j-1}}{\vartriangle S}}{\vartriangle S}  = \frac{f_{i,j+1} - 2f_{i,j} + f_{i,j-1}}{(\vartriangle S)^{2}}$

# 6.1 Roger & Shi's PDE

**6.1.1 IMPLICIT METHOD** : central in z, forward in t
* $\frac{\partial f}{\partial z} \approx \frac{f_{i,j+1} - f_{i,j-1}}{2 \vartriangle z}$
* $\frac{\partial f}{\partial t} \approx \frac{f_{i+1,j} - f_{i,j}}{\vartriangle t}$
* $\frac{\partial^{2}f}{\partial S^{2}} \approx  \frac{f_{i,j+1} - 2f_{i,j} + f_{i,j-1}}{(\vartriangle S)^{2}}$

where $z = z_0 + j(\vartriangle z)$ and $t = i (\vartriangle t)$

substituting for the partial derivatives in (2), we get:

$$\frac{f_{i+1,j} - f_{i,j}}{\vartriangle t} + \left( \frac{1}{T} - rz \right) \left( \frac{f_{i,j+1} - f_{i,j-1}}{2 \vartriangle z} \right) + \frac{1}{2} \sigma^{2}z^{2} \left( \frac{f_{i,j+1} - 2f_{i,j} + f_{i,j-1}}{(\vartriangle z)^{2}} \right) = 0$$

rearranging, we get 

$$f_{i+1,j} = a_j f_{i,j-1} + b_j f_{i,j} + c_j f_{i,j+1}$$

where $$a_j = \left( \frac{1-rzT}{2T \vartriangle z} - \frac{\sigma^{2} z^{2}}{2 (\vartriangle z)^{2}} \right) \vartriangle t$$
$$b_j = 1 + \frac{\sigma^{2} z^{2} \vartriangle t}{(\vartriangle z)^{2}}$$
$$c_j = \left( \frac{1-rzT}{2T \vartriangle z} + \frac{\sigma^{2} z^{2}}{2 (\vartriangle z)^{2}} \right) \vartriangle t$$

we have the boundary conditions: $f_{i,0} = 0$ and $f_{i,M} = z_M + \frac{t}{T}$

thus we get the system of equations:
$$f_{i+1,1} = a_1f_{i,0} + b_1f_{i,1} + c_1f_{i,2}$$
$$f_{i+1,2} = a_2f_{i,1} + b_2f_{i,2} + c_2f_{i,3}$$
$$\vdots$$
$$f_{i+1,M-1} = a_{M-1}f_{i,M-2} + b_{M-1}f_{i,M-1} + c_{M-1} f_{i,M}$$

thus we solve the following system as: 
$$\left[
  \begin{matrix}
b_1 & c_1 & & & & & \\
a_2 & b_2 & c_2 & & & & \\
& a_3 & b_3 & c_3 & & & \\
\vdots & \vdots & \vdots & \dots & &  \\
& & & & & & & a_{M-1} & b_{M-1} \\ 
\end{matrix}
\right] %
\left[
  \begin{matrix}
  f(i,1) \\
  f(i,2) \\
  f(i,3) \\
  \vdots \\
  f(i, M-1)
  \end{matrix}
  \right]= \left[
  \begin{matrix}
  f(i+1,1) - a_1f(i,0) \\
  f(i+1,2) \\
  f(i+1,3) \\
  \vdots \\
  f(i+1, M-1) - c_{M-1}f(i,M) 
  \end{matrix}
  \right]$$


```python
%matplotlib inline
import numpy as np
import scipy as sp
import scipy.sparse as spsp
import scipy.sparse.linalg as spsplg
import matplotlib.pyplot as plt


def RS_Implicit(S,K,r,sigma,T,graph=True,c=False):
    year_frac = 5000
    dt = 1/year_frac
    sim_points = int(year_frac *T)

    delta_z = 0.05
    upper_z = 0
    lower_z = -5

    n_grid = int((upper_z - lower_z)/delta_z) - 1
    if graph==True:
        print("n_grid:", n_grid)
    asset_price = np.linspace(lower_z + delta_z, upper_z - delta_z, n_grid)
    po_call = np.maximum(asset_price, 0)

    #implicit method
    def aj(j,r,dt,dz,sigma,T):
        z = lower_z + j*dz
        Aj = (0.5*((1-T*r*z)/(T*dz)) - 0.5*(sigma*z/dz)**2)*dt
        return Aj

    def bj(j,r,dt,dz,sigma,T):
        z = lower_z + j*dz
        Bj = 1 + (dt*(sigma*z/dz)**2)
        return Bj

    def cj(j,r,dt,dz,sigma,T):
        z = lower_z + j*dz
        Cj = -((0.5*(1-r*z*T)/(T*dz) + 0.5*(sigma*z/dz)**2))*dt
        return Cj


    # using sparse matrix methods:

    spg_mat = spsp.diags(
        [[aj(j,r,dt,delta_z,sigma,T) for j in range(2,n_grid+1)],
         [bj(j,r,dt,delta_z,sigma,T) for j in range(1,n_grid+1)],
         [cj(j,r,dt,delta_z,sigma,T) for j in range(1,n_grid)]]
    , [-1,0,1], (n_grid, n_grid)).tocsc()

    imRes = po_call.copy()
    a_1 = aj(1,r,dt,delta_z, sigma,T)
    c_m_1 = cj(n_grid, r, dt, delta_z,sigma,T)
    
    
    for i in range(sim_points-1):
        t = (i+1)*dt
        boundary1 = (1-np.exp(-r*t))/(r*T)
        
        imRes[0] -= a_1 * 0
        imRes[-1] -= c_m_1 *(upper_z + boundary1)
        imRes = spsplg.spsolve(spg_mat, imRes)

    lab="S = "+str(S)
    

    rsKlist = np.round(-S*(asset_price), 10)
    
    if graph==True:
        plt.plot(-S*(asset_price), imRes*S, label = "Roger & Shi")
        plt.show()
        
    if c == False:
        return (imRes*S)[np.where(rsKlist==K)][0]
    else:
        strikepdelist = -S*(asset_price)
        PDEpricelist = imRes*S
        strike = K
        upper_index = np.max(np.where(strikepdelist>strike))
        lower_index = np.min(np.where(strikepdelist<strike))
        upper_strike = strikepdelist[upper_index]
        lower_strike = strikepdelist[lower_index]
        upper_price = PDEpricelist[upper_index]
        lower_price = PDEpricelist[lower_index]
        price = ((upper_strike-strike)*lower_price/(upper_strike-lower_strike)+(strike-lower_strike)*upper_price/(upper_strike-lower_strike))
    
        return price

#plt.figure()
print("The option price using implicit method is:",RS_Implicit(S=100,K=95,r=0.02,sigma=0.5,T=1,graph = True,c=True))
```

**6.1.2 EXPLICIT METHOD**: symmetric in z, forward in t

$\frac{\partial f}{\partial z} = \frac{f(i+1,j+1)-f(i+1,j-1)}{2\Delta z}$

$\frac{\partial f}{\partial t} = \frac{f(i+1,j) - f(i,j)}{\Delta t}$

$\frac{\partial^2 f}{\partial z^2} = \frac{f(i+1,j+1)-2f(i+1,j)+f(i+1,j-1)}{\Delta z^2}$

substituting for partial derivatives in (2), we get:
$$\left(\frac{1}{T} - rz \right) \left(\frac{f(i+1,j) - f(i,j)}{\Delta t} \right) + \frac{f(i+1,j+1)-f(i+1,j-1)}{2\Delta z} +$$
$$\frac{1}{2}z^2\sigma^2 \left(\frac{f(i+1,j+1)-2f(i+1,j)+f(i+1,j-1)}{\Delta z^2} \right)$$

we get $f(i,j) = a_jf(i+1,j-1) +b_jf(i+1,j) + c_jf(i+1,j+1)$

$a_j = \Delta t \left(\frac{z^2\sigma^2}{2\Delta z^2} - \frac{1}{2\Delta z} \left(\frac{1}{T} - rz \right) \right)$

$b_j = \left(1 - \frac{z^2\sigma^2\Delta t}{\Delta z^2} \right)$

$c_j = \Delta t \left(\frac{z^2\sigma^2}{2\Delta z^2} + \frac{1}{2\Delta z} \left(\frac{1}{T} - rz \right) \right)$

We get the system of equations:

$$f_{i,1} = a_1f_{i+1,0} + b_1f_{i+1,1} + c_1f_{i+1,2}$$
$$f_{i,2} = a_1f_{i+1,1} + b_1f_{i+1,2} + c_1f_{i+1,3}$$
$$\vdots$$
$$f_{i,M-1} = a_1f_{i+1,M-2} + b_1f_{i+1,M-1} + c_1f_{i+1,M}$$

thus we solve the following system as:

$$\left[
  \begin{matrix}
b_1 & c_1 & & & & & \\
a_2 & b_2 & c_2 & & & & \\
& a_3 & b_3 & c_3 & & & \\
\vdots & \vdots & \vdots & \dots & &  \\
& & & & & & & a_{M-1} & b_{M-1} \\ 
\end{matrix}
\right] %
\left[
  \begin{matrix}
  f(i+1,1) \\
  f(i+1,2) \\
  f(i+1,3) \\
  \vdots \\
  f(i+1, M-1)
  \end{matrix}
  \right]= \left[
  \begin{matrix}
  f(i,1) - a_1f(i+1,0) \\
  f(i,2) \\
  f(i,3) \\
  \vdots \\
  f(i, M-1) - c_{M-1}f(i+1,M) 
  \end{matrix}
  \right]$$


```python
def RS_Explicit(S,K,r,sigma,T,graph=True):
    year_frac = 5000
    dt = 1/year_frac
    sim_points = int(year_frac *T)

    delta_z = 0.05
    upper_z = 0
    lower_z = -5

    n_grid = int((upper_z - lower_z)/delta_z) - 1
    if graph==True:
        print("n_grid:", n_grid)
    asset_price = np.linspace(lower_z + delta_z, upper_z - delta_z, n_grid)
    po_call = np.maximum(asset_price, 0)

    def aj(j,r,dt,dz,sigma,T):
        z = lower_z + j*dz
        aj = dt*(0.5*(sigma*z/dz)**2 - (1/(2*dz))*((1/T)-r*z))
        return aj

    def bj(j,r,dt,dz,sigma,T):
        z = lower_z + j*dz
        bj = 1 - dt*(z*sigma/dz)**2
        return bj

    def cj(j,r,dt,dz,sigma,T):
        z = lower_z + j*dz
        cj = dt*(0.5*(sigma*z/dz)**2 + (1/(2*dz))*((1/T)-r*z))
        return cj

    res = po_call.copy()
    res = res.reshape((n_grid,1))

    a_1 = aj(1,r,dt,delta_z,sigma,T)
    c_m = cj(n_grid,r,dt,delta_z,sigma,T)

    explicitmat = spsp.diags(
        [[aj(j,r,dt,delta_z,sigma,T) for j in range(2,n_grid+1)],
         [bj(j,r,dt,delta_z,sigma,T) for j in range(1,n_grid+1)],
         [cj(j,r,dt,delta_z,sigma,T) for j in range(1,n_grid)]]
    , [-1,0,1], (n_grid, n_grid)).todense()

    #print(type(explicitmat),type(res))
    #print(res.shape)
    #print(spsp.csr_matrix.get_shape(explicitmat),spsp.csr_matrix.get_shape(res))

    for i in range(sim_points-1):
        t = (i+1)*dt
        res = np.matmul(explicitmat, res)
        res[0] += a_1 * 0
        res[-1] += c_m*(upper_z +t/T)

    rsKlist = np.round(-S*(asset_price),3)
    rsCN = (res*S).copy()
    if graph==True:

        plt.plot(-S*(asset_price), res*S, label = 'Roger & Shi')
        plt.show()
        
    
    return (res*S)[np.where(rsKlist==K)][0][0] 

plt.figure()
print("The option price using explicit scheme is:",RS_Explicit(S=100,K=110,r=0.02,sigma=0.5,T=0.05, graph = True)[0][0])
```

**6.1.3 CRANK NICOLSON METHOD**: average of explicit and implicit scheme

* $\frac{\partial f}{\partial z} = \frac{1}{2} \left( \frac{f_{i,j+1} - f_{i,j-1}}{2 \vartriangle z} + \frac{f_{i-1,j+1} - f_{i-1,j-1}}{2 \vartriangle z} \right)$
* $\frac{\partial f}{\partial t} = \frac{f_{i,j}-f_{i-1,j}}{\vartriangle t}$
* $\frac{\partial ^{2} f}{\partial z^{2}} = \frac{1}{2} \left( \frac{f_{i,j+1} - 2f_{i,j} + f_{i,j-1}}{(\vartriangle z)^{2}} + \frac{f_{i-1,j+1} - 2f_{i-1,j} + f_{i-1,j-1}}{(\vartriangle z)^{2}} \right)$

substituting for partial derivatives in (2), we get:
$$\frac{f_{i,j} - f_{i-1,j}}{\vartriangle t} + \left( \frac{1}{T} - rz \right) \left( \frac{1}{2} \left( \frac{f_{i,j+1} - f_{i,j-1}}{2 \vartriangle z} + \frac{f_{i-1,j+1} -f_{i-1,j-1}}{2\vartriangle z} \right) \right) +$$ $$\frac{1}{4}z^{2}\sigma^{2}\left( \frac{f_{i,j+1} -2f_{i,j} +f_{i-1,j-1}}{(\vartriangle z)^{2}} + \frac{f_{i-1,j+1} -2f_{i-1,j} + f_{i-1,j-1}}{(\vartriangle z)^{2}} \right) = 0$$

rearranging, we get $a_j f_{i,j-1} + (1-b_j)f_{i,j} + c_jf_{i,j+1} = -a_j f_{i-1,j-1} + (1+b_j)f_{i-1,j} - c_j f_{i-1,j+1}$

where 
$$a_j = \left( \frac{z^{2}\sigma^{2}}{4(\vartriangle z)^{2}} - \frac{1-rzT}{4T \vartriangle z} \right) \vartriangle t$$
$$b_j = \frac{\sigma^{2}z^{2} \vartriangle t}{2 (\vartriangle z)^{2}}$$
$$c_j = \left( \frac{z^{2}\sigma^{2}}{4(\vartriangle z)^{2}} + \frac{1-rzT}{4T \vartriangle z} \right) \vartriangle t$$

we get the system of equations:
$$a_1 f_{i,0} + (1-b_1)f_{i,1} + c_1f_{i,2} = -a_1f_{i-1,0} + (1+b_1)f_{i-1,1} - c_1f_{i-1,2}$$
$$a_2 f_{i,1} + (1-b_2)f_{i,2} + c_2 f_{i,3} = -a_2f_{i-1,1} + (1+b_2)f_{i-1,2}  - c_2 f_{i-1,3}$$
$$\vdots$$
$$a_{M-1} f_{i,M-2} + (1-b_{M-1})f_{i,M-1} + c_{M-1}f_{i,M} = -a_{M-1}f_{i-1,M-2} + (1+b_{M-1})f_{i-1,M-1} - c_{M-1}f_{i-1,M}$$

we solve the system as follows:

$$\left[
  \begin{matrix}
(1-b_1) & c_1 & & & & & \\
a_2 & (1-b_2) & c_2 & & & & \\
& a_3 & (1-b_3) & c_3 & & & \\
\vdots & \vdots & \vdots & \dots & &  \\
& & & & & & & a_{M-1} & (1-b_{M-1}) \\ 
\end{matrix}
\right] %
\left[
  \begin{matrix}
  f(i,1) \\
  f(i,2) \\
  f(i,3) \\
  \vdots \\
  f(i, M-1)
  \end{matrix}
  \right]=$$
  $$\left[
  \begin{matrix}
(1+b_1) & -c_1 & & & & & \\
-a_2 & (1+b_2) & -c_2 & & & & \\
& -a_3 & (1+b_3) & -c_3 & & & \\
\vdots & \vdots & \vdots & \dots & &  \\
& & & & & & & -a_{M-1} & (1+b_{M-1}) \\ 
\end{matrix}
\right] %
\left[
  \begin{matrix}
  f(i-1,1) \\
  f(i-1,2) \\
  f(i-1,3) \\
  \vdots \\
  f(i-1, M-1)
  \end{matrix}
  \right]$$



```python
def RS_CN(S,K,r,sigma,T,graph=True,c=False):
    year_frac = 5000
    dt = 1/year_frac
    sim_points = int(year_frac *T)

    delta_z = 0.05
    upper_z = 0
    lower_z = -5
    
    n_grid = int((upper_z - lower_z)/delta_z) - 1
    asset_price = np.linspace(lower_z + delta_z, upper_z - delta_z, n_grid)
    po_call = np.maximum(asset_price, 0)
    res = po_call.copy()
    res.reshape((n_grid,1))
    #crank nicolson method

    def cnaj(j, r, dt, dz, sigma, T):
        z = lower_z + j*dz
        cnaj = ((0.25*(z*sigma/dz)**2) - ((1-T*r*z)/(4*T*dz)))*dt
        return cnaj

    def cnbj(j,r,dt,dz,sigma,T):
        z = lower_z + j*dz
        cnbj = 0.5*dt*(sigma*z/dz)**2
        return cnbj

    def cncj(j,r,dt,dz,sigma,T):
        z = lower_z + j*dz
        cncj = dt*((0.25*(z*sigma/dz)**2)+ ((1-r*z*T)/(4*T*dz)))
        return cncj

    cn_mat1 = spsp.diags([[cnaj(j,r,dt, delta_z, sigma,T) for j in range(2,n_grid+1)],
                          [1-cnbj(j,r,dt,delta_z, sigma, T) for j in range(1,n_grid+1)],
                          [cncj(j,r,dt,delta_z, sigma, T) for j in range(1, n_grid)]], [-1,0,1], (n_grid, n_grid)).todense()

    cn_mat2 = spsp.diags([[-cnaj(j,r,dt, delta_z, sigma,T) for j in range(2,n_grid+1)],
                          [1+cnbj(j,r,dt,delta_z, sigma, T) for j in range(1,n_grid+1)],
                          [-cncj(j,r,dt,delta_z, sigma, T) for j in range(1, n_grid)]], [-1,0,1], (n_grid, n_grid)).tocsc()

    cna_1 = cnaj(1,r,dt,delta_z, sigma, T)
    cnc_m = cncj(n_grid,r,dt,delta_z, sigma, T)

    for i in range(sim_points-1):
        t = (i+1)*dt
        eq1 = np.matmul(cn_mat1, res)
        eq1[0] += cna_1 * 0 + cna_1 * 0
        eq1 = eq1.reshape((n_grid, 1))
        eq1[-1] += cnc_m*(upper_z + t/T) + cnc_m*(upper_z + (t+dt)/T)
        res = spsplg.spsolve(cn_mat2, eq1)

    rsKlist = np.round(-S*(asset_price),3)
  
    if graph==True:
        print("n_grid: ",n_grid)
        plt.plot(-S*(asset_price), res*S,label = 'Roger & Shi')
    
    if c == False:
        return (res*S)[np.where(rsKlist==K)][0]
    else:
        strikepdelist = -S*(asset_price)
        PDEpricelist = res*S
        strike = K
        upper_index = np.max(np.where(strikepdelist>strike))
        lower_index = np.min(np.where(strikepdelist<strike))
        upper_strike = strikepdelist[upper_index]
        lower_strike = strikepdelist[lower_index]
        upper_price = PDEpricelist[upper_index]
        lower_price = PDEpricelist[lower_index]
        price = ((upper_strike-strike)*lower_price/(upper_strike-lower_strike)+(strike-lower_strike)*upper_price/(upper_strike-lower_strike))
    
        return price
    

print("The option price using Crank Nicolson is:", RS_CN(S=100,K=95,r=0.02,sigma=2,T=1,graph = True,c=True))
```

```python

def Vecer_Implicit(S,K,r,sigma,T,graph = True,c=False):
    year_frac = 199
    dt = 1/year_frac
    sim_points = int(year_frac*T)

    delta_z = 0.01
    upper_z = 1
    lower_z = -1

    n_grid = int((upper_z - lower_z)/delta_z) - 1
    if graph==True:
        print("n_grid:", n_grid)
    asset_price = np.linspace(lower_z + delta_z, upper_z - delta_z, n_grid)
    po_call = np.maximum(asset_price, 0)

    def aj(j,r,dt,dz,sigma,T,qt):
        z = lower_z + (j*dz)
        Aj = -(0.5*(sigma*(qt-z)/dz)**2 - (0.5*r*(qt-z)/dz))*dt
        return Aj

    def bj(j,r,dt,dz,sigma,T,qt):
        z = lower_z + (j*dz)
        Bj = 1 + dt*((sigma*(qt-z))/dz)**2
        return Bj

    def cj(j,r,dt,dz,sigma,T,qt):
        z = lower_z + (j*dz)
        Cj = -(0.5*(sigma*(qt-z)/dz)**2 + (0.5*r*(qt-z)/dz))*dt
        return Cj

   
    res = po_call.copy()
    res.reshape((n_grid,1))
    
    tlist = []
    for i in range(sim_points):
        t = (i+1)*dt
        qt = 1-(t/T)
        tlist.append(qt)
        
        
        spg_mat = spsp.diags(
        [[aj(j,r,dt,delta_z,sigma,T,qt) for j in range(2,n_grid+1)],
         [bj(j,r,dt,delta_z,sigma,T,qt) for j in range(1,n_grid+1)],
         [cj(j,r,dt,delta_z,sigma,T,qt) for j in range(1,n_grid)]]
        ,[-1,0,1], (n_grid, n_grid)).tocsc()

        a_1 = aj(1,r,dt,delta_z, sigma,T,qt)
        c_m_1 = cj(n_grid, r, dt, delta_z,sigma,T,qt)
        
        res[0] -= 0
        res[-1] -= c_m_1*(2*res[-1] - res[-2])*np.exp(-r*dt)
        res = spsplg.spsolve(spg_mat,res)
        
    if graph == True:
        plt.plot(S*(1-asset_price), res*S, label = "Vecer")   
    Klist = np.round((1-asset_price)*S, 3)
    
    if c == False:
        return (res*S)[np.where(Klist==K)][0]
    else:
        strikepdelist = S*(1-asset_price)
        PDEpricelist = res*S
        strike = K
        upper_index = np.max(np.where(strikepdelist>strike))
        lower_index = np.min(np.where(strikepdelist<strike))
        upper_strike = strikepdelist[upper_index]
        lower_strike = strikepdelist[lower_index]
        upper_price = PDEpricelist[upper_index]
        lower_price = PDEpricelist[lower_index]
        price = ((upper_strike-strike)*lower_price/(upper_strike-lower_strike)+(strike-lower_strike)*upper_price/(upper_strike-lower_strike))
    
        return price
    
print("The option price using implicit method is:",Vecer_Implicit(S=100,K=95,r=0.02,sigma=0.5,T=1,graph = True,c=True))
```

**6.2.3 CRANK NICOLSON METHOD**

we have:

* $\frac{\partial u}{\partial t} = \frac{f_{i,j} - f_{i-1,j}}{\vartriangle t}$
* $\frac{\partial u}{\partial z} = \frac{1}{2} \left( \frac{f_{i,j+1} - f_{i,j-1}}{2 \vartriangle z} + \frac{f_{i-1,j+1} - f_{i-1,j-1}}{2 \vartriangle z} \right)$
* $\frac{\partial^{2} u}{\partial z^{2}} = \frac{1}{2} \left( \frac{f_{i,j+1} - 2f_{i,j} + f_{i,j-1}}{(\vartriangle z)^{2}} + \frac{f_{i-1,j+1} - 2f_{i-1,j} + f_{i-1,j-1}}{(\vartriangle z)^{2}} \right)$

substituting for the partial derivatives in (3), we get:
$$\frac{f_{i,j} - f_{i-1,j}}{\vartriangle t} + r\frac{(q_t -z)}{2} \left( \frac{f_{i,j+1} - f_{i,j-1}}{2 \vartriangle z} + \frac{f_{i-1,j+1} - f_{i-1,j-1}}{2 \vartriangle z} \right) +$$ $$\frac{\sigma^{2}(q_t -z)^{2}}{4} \left( \frac{f_{i,j+1} - 2f_{i,j} + f_{i,j-1}}{(\vartriangle z)^{2}} + \frac{f_{i-1,j+1} - 2f_{i-1,j} + f_{i-1,j-1}}{(\vartriangle z)^{2}} \right)$$

rearranging, we may write this as :
$$a_j f_{i,j-1} + (1 -b_j)f_{i,j} + c_jf_{i,j+1} = -a_j f_{i-1,j-1} + (1 + b_j)f_{i-1,j} - c_jf_{i-1,j+1}$$
 where $a_j = \left( \frac{\sigma^{2} (q_t -z)^{2}}{4(\vartriangle z)^{2}} - \frac{r(q_t -z)}{4 \vartriangle z} \right) \vartriangle t$

$b_j = \frac{\sigma^{2}(q_t -z)^{2}}{2 (\vartriangle z )^{2}}$

$c_j = \left( \frac{\sigma^{2} (q_t -z)^{2}}{4(\vartriangle z)^{2}} + \frac{r(q_t -z)}{4 \vartriangle z} \right) \vartriangle t$

Thus we get the following system of equations:
$$a_1f_{i,0} + (1-b_1)f_{i,1} + c_1 f_{i,2} = -a_1 f_{i-1,0} + (1-b_1)f_{i-1,1} -c_1f_{i-1,2}$$
$$a_2f_{i,1} + (1-b_2)f_{i,2} + c_2 f_{i,3} = -a_2 f_{i-1,1} + (1-b_2)f_{i-1,2} -c_2f_{i-1,3}$$
$$\vdots$$
$$a_{M-1}f_{i,M-2} + (1-b_{M-1})f_{i,M-1} + c_{M-1} f_{i,M} = -a_{M-1} f_{i-1,M-2} + (1-b_{M-1})f_{i-1,M-1} -c_{M-1}f_{i-1,M}$$ 

with boundary conditions $f_{i,0} = 0$ and $f_{N,j} = 2f_{M-1,j} - 2f_{M-2,j}$

we solve the system as follows:
$$\left[
  \begin{matrix}
(1-b_1) & c_1 & & & & & \\
a_2 & (1-b_2) & c_2 & & & & \\
& a_3 & (1-b_3) & c_3 & & & \\
\vdots & \vdots & \vdots & \dots & &  \\
& & & & & & & a_{M-1} & (1-b_{M-1}) \\ 
\end{matrix}
\right] %
\left[
  \begin{matrix}
  f(i,1) \\
  f(i,2) \\
  f(i,3) \\
  \vdots \\
  f(i, M-1)
  \end{matrix}
  \right]=$$
  $$\left[
  \begin{matrix}
  (1+b_1) & -c_1 & & & & & \\
  -a_2 & (1+b_2) & -c_2 & & & & \\
  & -a_3 & (1+b_3) & -c_3 & & & \\
  \vdots & \vdots & \vdots & \dots & &  \\
  & & & & & & & -a_{M-1} & (1+b_{M-1}) \\ 
  \end{matrix}
  \right] %
  \left[
  \begin{matrix}
  f(i-1,1) \\
  f(i-1,2) \\
  f(i-1,3) \\
  \vdots \\
  f(i-1, M-1)
  \end{matrix}
  \right]$$

```python
import numpy as np
import scipy as sp
import scipy.sparse as spsp
import scipy.sparse.linalg as spsplg
import matplotlib.pyplot as plt

def Vecer_CN(S,K,r,sigma,T,graph = True,c=False):
    year_frac = 5000
    dt = 1/year_frac
    sim_points = int(year_frac*T)

    delta_z = 0.01
    upper_z = 1
    lower_z = -1

    n_grid = int((upper_z - lower_z)/delta_z)-1
    asset_price = np.linspace(lower_z + delta_z, upper_z - delta_z, n_grid)
    po_call = np.maximum(asset_price, 0)
    if graph == True:
        print("n_grid:", n_grid)

    def aj(j,r,dt,dz,sigma,T,qt):
        z = lower_z + j*dz
        aj = (0.25*(sigma*(qt - z)/dz)**2 - 0.25*(r*(qt-z)/dz))*dt
        return aj

    def bj(j,r,dt,dz,sigma,T,qt):
        z = lower_z + j*dz
        bj = 0.5*dt*(sigma*(qt-z)/dz)**2
        return bj

    def cj(j,r,dt,dz,sigma,T,qt):
        z = lower_z + j*dz
        cj = (0.25*(sigma*(qt - z)/dz)**2 + 0.25*(r*(qt-z)/dz))*dt
        return cj

    res = po_call.copy()
    res = res.reshape((n_grid,1))

    for i in range(sim_points):
        t = i*dt
        qt = 1-(t/T)
        a1 = aj(1,r,dt,delta_z,sigma,T,qt)
        c_m = cj(n_grid,r,dt,delta_z,sigma,T,qt)

        cnmat1 = spsp.diags(
            [[aj(j,r,dt,delta_z,sigma,T,qt) for j in range(2, n_grid+1)],
            [1-bj(j,r,dt,delta_z,sigma,T,qt) for j in range(1,n_grid+1)],
            [cj(j,r,dt,delta_z,sigma,T,qt) for j in range(1,n_grid)]],
            [-1,0,1], (n_grid, n_grid)).todense()
  
        cnmat2 = spsp.diags( 
            [[-aj(j,r,dt,delta_z,sigma,T,qt) for j in range(2,n_grid+1)],
            [1+bj(j,r,dt,delta_z,sigma,T,qt) for j in range(1,n_grid+1)],
            [-cj(j,r,dt,delta_z, sigma, T,qt) for j in range(1,n_grid)]],
            [-1,0,1], (n_grid, n_grid)).tocsc()

        eq1 = np.matmul(cnmat1,res)
        eq1[0] += 0 
        eq1[-1] += c_m*(2*res[-1] - res[-2]) + c_m*(2*res[-1]-res[-2])
        eq1 = eq1.reshape((n_grid, 1))
        res = spsplg.spsolve(cnmat2, eq1)
        res = res.reshape((n_grid,1))

  
    Klist = np.round((1-asset_price)*S, 3)
    CN = (res*S).copy()
    
    if graph == True:
        plt.plot((1-asset_price)*S, res*S, label = "Vecer")
    
    if c == False:
        return (res*S)[np.where(Klist==K)][0]
    else:
        strikepdelist = S*(1-asset_price)
        PDEpricelist = res*S
        strike = K
        upper_index = np.max(np.where(strikepdelist>strike))
        lower_index = np.min(np.where(strikepdelist<strike))
        upper_strike = strikepdelist[upper_index]
        lower_strike = strikepdelist[lower_index]
        upper_price = PDEpricelist[upper_index]
        lower_price = PDEpricelist[lower_index]
        price = ((upper_strike-strike)*lower_price/(upper_strike-lower_strike)+(strike-lower_strike)*upper_price/(upper_strike-lower_strike))[0]
    
        return price

print("The option price using crank nicolson is:",Vecer_CN(S=100,K=90,r=0.02,sigma=2,T=1,graph = True,c=True))
```


**6.3.3 CRANK NICOLSON METHOD**

Note that by our derivation in Section 5.3 we note that the numerical approximation framework for Dubois-Lelievre will be similar to that of Vecer's PDE but we replace stock position $q_t = 1 - \frac{t}{T}$ to $\frac{t}{T}$.

we may write this as :
$$a_j f_{i,j-1} + (1 -b_j)f_{i,j} + c_jf_{i,j+1} = -a_j f_{i-1,j-1} + (1 + b_j)f_{i-1,j} - c_jf_{i-1,j+1}$$
 where $a_j = \left( \frac{\sigma^{2} (q_t -z)^{2}}{4(\vartriangle z)^{2}} - \frac{r(q_t -z)}{4 \vartriangle z} \right) \vartriangle t$

$b_j = \frac{\sigma^{2}(q_t -z)^{2}}{2 (\vartriangle z )^{2}}$

$c_j = \left( \frac{\sigma^{2} (q_t -z)^{2}}{4(\vartriangle z)^{2}} + \frac{r(q_t -z)}{4 \vartriangle z} \right) \vartriangle t$

Thus we get the following system of equations:
$$a_1f_{i,0} + (1-b_1)f_{i,1} + c_1 f_{i,2} = -a_1 f_{i-1,0} + (1-b_1)f_{i-1,1} -c_1f_{i-1,2}$$
$$a_2f_{i,1} + (1-b_2)f_{i,2} + c_2 f_{i,3} = -a_2 f_{i-1,1} + (1-b_2)f_{i-1,2} -c_2f_{i-1,3}$$
$$\vdots$$
$$a_{M-1}f_{i,M-2} + (1-b_{M-1})f_{i,M-1} + c_{M-1} f_{i,M} = -a_{M-1} f_{i-1,M-2} + (1-b_{M-1})f_{i-1,M-1} -c_{M-1}f_{i-1,M}$$ 

with boundary conditions $f_{i,0} = 0$ and $f_{N,j} = 2f_{M-1,j} - 2f_{M-2,j}$

we solve the system as follows:
$$\left[
  \begin{matrix}
(1-b_1) & c_1 & & & & & \\
a_2 & (1-b_2) & c_2 & & & & \\
& a_3 & (1-b_3) & c_3 & & & \\
\vdots & \vdots & \vdots & \dots & &  \\
& & & & & & & a_{M-1} & (1-b_{M-1}) \\ 
\end{matrix}
\right] %
\left[
  \begin{matrix}
  f(i,1) \\
  f(i,2) \\
  f(i,3) \\
  \vdots \\
  f(i, M-1)
  \end{matrix}
  \right]=$$
  $$\left[
  \begin{matrix}
  (1+b_1) & -c_1 & & & & & \\
  -a_2 & (1+b_2) & -c_2 & & & & \\
  & -a_3 & (1+b_3) & -c_3 & & & \\
  \vdots & \vdots & \vdots & \dots & &  \\
  & & & & & & & -a_{M-1} & (1+b_{M-1}) \\ 
  \end{matrix}
  \right] %
  \left[
  \begin{matrix}
  f(i-1,1) \\
  f(i-1,2) \\
  f(i-1,3) \\
  \vdots \\
  f(i-1, M-1)
  \end{matrix}
  \right]$$


```python
def DL_CN(S,K,r,sigma,T,graph = True,c=False):
    year_frac = 5000
    dt = 1/year_frac
    sim_points = int(year_frac*T)

    delta_z = 0.01
    upper_z = 1
    lower_z = -1

    n_grid = int((upper_z - lower_z)/delta_z)-1
    asset_price = np.linspace(lower_z + delta_z, upper_z - delta_z, n_grid)
    po_call = np.maximum(asset_price, 0)
    if graph == True:
        print("n_grid:", n_grid)

    def aj(j,r,dt,dz,sigma,T,qt):
        z = lower_z + j*dz
        aj = (0.25*(sigma*(qt - z)/dz)**2 - 0.25*(r*(qt-z)/dz))*dt
        return aj

    def bj(j,r,dt,dz,sigma,T,qt):
        z = lower_z + j*dz
        bj = 0.5*dt*(sigma*(qt-z)/dz)**2
        return bj

    def cj(j,r,dt,dz,sigma,T,qt):
        z = lower_z + j*dz
        cj = (0.25*(sigma*(qt - z)/dz)**2 + 0.25*(r*(qt-z)/dz))*dt
        return cj

    res = po_call.copy()
    res = res.reshape((n_grid,1))

    for i in range(sim_points):
        t = i*dt
        qt =(t/T)
        a1 = aj(1,r,dt,delta_z,sigma,T,qt)
        c_m = cj(n_grid,r,dt,delta_z,sigma,T,qt)

        cnmat1 = spsp.diags(
            [[aj(j,r,dt,delta_z,sigma,T,qt) for j in range(2, n_grid+1)],
            [1-bj(j,r,dt,delta_z,sigma,T,qt) for j in range(1,n_grid+1)],
            [cj(j,r,dt,delta_z,sigma,T,qt) for j in range(1,n_grid)]],
            [-1,0,1], (n_grid, n_grid)).todense()
  
        cnmat2 = spsp.diags( 
            [[-aj(j,r,dt,delta_z,sigma,T,qt) for j in range(2,n_grid+1)],
            [1+bj(j,r,dt,delta_z,sigma,T,qt) for j in range(1,n_grid+1)],
            [-cj(j,r,dt,delta_z, sigma, T,qt) for j in range(1,n_grid)]],
            [-1,0,1], (n_grid, n_grid)).tocsc()

        eq1 = np.matmul(cnmat1,res)
        eq1[0] += 0 
        eq1[-1] += c_m*(2*res[-1] - res[-2]) + c_m*(2*res[-1]-res[-2])
        eq1 = eq1.reshape((n_grid, 1))
        res = spsplg.spsolve(cnmat2, eq1)
        res = res.reshape((n_grid,1))

  
    Klist = np.round((1-asset_price)*S, 3)
    CN = (res*S).copy()

    if graph == True:
        plt.plot((1-asset_price)*S, res*S, label = "Dubois Lelievre")
    
    if c == False:
        return (res*S)[np.where(Klist==K)][0]
    else:
        strikepdelist = S*(1-asset_price)
        PDEpricelist = res*S
        strike = K
        upper_index = np.max(np.where(strikepdelist>strike))
        lower_index = np.min(np.where(strikepdelist<strike))
        upper_strike = strikepdelist[upper_index]
        lower_strike = strikepdelist[lower_index]
        upper_price = PDEpricelist[upper_index]
        lower_price = PDEpricelist[lower_index]
        price = ((upper_strike-strike)*lower_price/(upper_strike-lower_strike)+(strike-lower_strike)*upper_price/(upper_strike-lower_strike))[0]
    
        return price

print("The option price using crank nicolson is:",DL_CN(S=100,K=95,r=0.02,sigma=2,T=1,graph = True, c=True))
```

# Lognormal Approximation 

```python
from numpy import exp
from scipy.stats import norm
import math
from math import log,sqrt

from scipy.stats import gamma as ga




def lognormalapprox(sigma,stock,strike,T,r):
    
    if(type(sigma)==np.ndarray):
        sigma = sigma[0]
        
    m1 = stock*(exp(r*T)-1)/r
    m2 = 2*stock**2*(r*exp((sigma**2+2*r)*T)-(sigma**2+2*r)*exp(r*T)+(sigma**2+r))/(sigma**2+r)/(sigma**2+2*r)/r
    eta = log(m2/m1**2)/T
    mu = log(m1)/T-0.5*eta
    d1 = (mu*T+eta*T-log(strike*T))/sqrt(eta*T)
    d2 = d1-sqrt(eta*T)
    return exp((mu+eta/2)*T)*norm.cdf(d1)/T-strike*norm.cdf(d2)


print("The option price generated by lognormal approximation is:",lognorm(0.5,100,95,1,0.02))




```

# Gamma Approximation

```python
from scipy.stats import gamma
def gammaapprox(stock,strike,T,r,sigma):
    m1 = stock*(exp(r*T)-1)/r/T
    m2 = 2*stock**2*(r*exp((sigma**2+2*r)*T)-(sigma**2+2*r)*exp(r*T)+(sigma**2+r))/(sigma**2+r)/(sigma**2+2*r)/r/T**2
    alpha = (2*m2-m1**2)/(m2-m1**2)
    beta = (m2-m1**2)/m2/m1
    return (1-exp(-r*T))*stock*ga.cdf(1/strike,a=alpha-1,scale=beta)/r/T-exp(-r*T)*strike*ga.cdf(1/strike,a=alpha,scale=beta)


print("The option price generated by gamma approximation is:",gammaapprox(100,105.1709,5,0.02,0.5))

```

# Standard Monte Carlo

```python
import numpy as np

S0 = 100
K = 110
r = 0.02
sigma = 0.5
T = 0.05
strike = K
rf = r
ttm = T
N = 100
n = 10000

def brownian(t,n):
    Bs = np.zeros(n)
    ds = t/n
    for i in range(1,n):
        sign = np.random.choice([-1,1])
        Bs[i] = Bs[i-1]+ sign*np.sqrt(ds)
    return Bs

W_t = np.random.randn(n,N)



def MCS(N,n,K,T,ifcum):
    delta_t = T/N
    pv = np.exp(-r*T)
    W = W_t[:n+1,:]
    zs = np.cumsum((r-0.5*sigma**2)*delta_t+sigma*np.sqrt(delta_t)*W,axis=1)
    S = S0*np.exp(zs)
    AverageS =  S.mean(axis=1)
        
    asianCallPayoff = np.maximum(AverageS-K,0)*pv
    asianCallPrice = asianCallPayoff.mean()
    
    return asianCallPrice,np.std(asianCallPayoff)/np.sqrt(n)

res = MCS(N,n,K,T,False)
print("The standard MC Asian Call price is:",res[0],", the standard error is:",res[1])  
```
